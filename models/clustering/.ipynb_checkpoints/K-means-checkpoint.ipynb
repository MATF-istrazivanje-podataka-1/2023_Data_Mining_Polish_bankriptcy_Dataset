# K-means

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from joblib import load
```

```python
X, Y = load('../../preprocessing/clustering_data.joblib')
```

```python
X.info()
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
reduced_X = pca.fit_transform(X)
```

```python
reduced_X
```

```python
explained_variance_ratio = pca.explained_variance_ratio_

for i, ratio in enumerate(explained_variance_ratio):
    print(f'Component {i + 1} variance : {ratio:.4f}')
    
cumulative_variance_ratio = explained_variance_ratio.cumsum()

print(f'Cumulative vaiance: {cumulative_variance_ratio[-1]:.4f}')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances_argmin_min

def apply_kmeans_and_plot_3d(data, n_clusters):
    # k-means i iscrtavanje 3d grafika za klastere
  
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)

    labels = kmeans.labels_

    cluster_centers = kmeans.cluster_centers_
    silhouette_avg = silhouette_score(data, labels)
    inertia = kmeans.inertia_
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    for cluster_label in range(n_clusters):
        cluster_data = data[labels == cluster_label]
        ax.scatter(cluster_data[:, 0], cluster_data[:, 1], cluster_data[:, 2], label=f'Cluster {cluster_label + 1}')
    
    ax.scatter(cluster_centers[:, 0], cluster_centers[:, 1], cluster_centers[:, 2], marker='X', s=200, c='black', label='Centroids')
    ax.legend()
    
    ax.set_title(f"K means for {n_clusters}")
    
    ax.set_xlabel('PCA 1 component')
    ax.set_ylabel('PCA 2 component')
    ax.set_zlabel('PCA 3 component')
    
    plt.show()
    
    return (silhouette_avg, inertia)
```

```python
n_clusters_list = [2, 3, 4, 5, 6, 7]

# Iterirajte kroz listu brojeva klastera i pozovite funkciju za svaki broj
sil_coefs = []
inertias = []

for n_clusters in n_clusters_list:
    s, i = apply_kmeans_and_plot_3d(reduced_X, n_clusters)
    sil_coefs.append(s)
    inertias.append(i)
```

Mozemo primetiit da se tacno odvaja 6 klastera. Provericemo i kvalitet klasterovanja pomocu koeficijenta senki i SSE.

```python
def plot_silhouette_scores(n_clusters_list, sil_coefs):
    """
    Iscrtava grafik za Silhouette koeficijente za različit broj klastera.

    """
    plt.figure(figsize=(10, 5))
    plt.plot(n_clusters_list, sil_coefs, marker='o', linestyle='-')
    plt.title('Silhouette score for different number of clusters')
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette score')
    plt.grid(True)
    plt.show()
```

```python
def plot_sse(n_clusters_list, inertias):
    """
    Iscrtava grafik za Sum of Squared Distances (SSE) u zavisnosti od broja klastera.

    """
    plt.figure(figsize=(10, 5))
    plt.plot(n_clusters_list, inertias, marker='o', linestyle='-')
    plt.title('SSE / Number of clusters')
    plt.xlabel('Number of clusters')
    plt.ylabel('SSE')
    plt.grid(True)
    plt.show()
```

```python
plot_silhouette_scores(n_clusters_list, sil_coefs)
```

```python
plot_sse(n_clusters_list, inertias)
```

Na osnovu grafika, pomocu metoda lakta i koeficijenta senki. Mozemo zakljuciti da je odgovarajuci broj klastera jednak 6.

### t-SNE

Primenicemo i drugu dimenzionu redukciju i pokusati ponovo da odredimo broj klastera.
t-SNE je metoda za vizualizaciju i smanjenje dimenzionalnosti koja se uglavnom koristi za očuvanje lokalnih struktura u podacima, dok PCA teži da očuva globalne varijanse.

```python
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt

n_components = 2  # Postavite na 2 za rad u dvodimenzionalnom prostoru
tsne = TSNE(n_components=n_components)

reduced_X_tsne = tsne.fit_transform(X)

# Iscrtavanje smanjenih podataka u 2D prostoru
plt.figure(figsize=(8, 6))
plt.scatter(reduced_X_tsne[:, 0], reduced_X_tsne[:, 1])
plt.title(f't-SNE with {n_components} components')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

Mozemo da primetimo da su se i ovde iskristalisale odredjene grupe, ovo moze biti posledica toga sto nas skup podataka sadrzi veliki broj kategorickih atributa.

```python
def apply_kmeans_and_plot_2d(data, n_clusters):
    # k-means i iscrtavanje 2D grafika za klastere
        
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)

    labels = kmeans.labels_
    cluster_centers = kmeans.cluster_centers_
    silhouette_avg = silhouette_score(data, labels)
    inertia = kmeans.inertia_
    
    plt.figure(figsize=(8, 6))
    
    for cluster_label in range(n_clusters):
        cluster_data = data[labels == cluster_label]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_label + 1}')
    
    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='X', s=200, c='black', label='Centroids')
    plt.legend()
    
    plt.title(f"K-Means for {n_clusters} Clusters")
    
    plt.xlabel('1 Component')
    plt.ylabel('2 Component')
    
    plt.grid(True)
    plt.show()
    
    return silhouette_avg, inertia
```

```python
n_clusters_list = [2, 4, 8, 16, 24]

# Iterirajte kroz listu brojeva klastera i pozovite funkciju za svaki broj
sil_coefs_tsne = []
inertias_tsne = []

for n_clusters in n_clusters_list:
    s, i = apply_kmeans_and_plot_2d(reduced_X_tsne, n_clusters)
    sil_coefs_tsne.append(s)
    inertias_tsne.append(i)
```

```python
plot_silhouette_scores(n_clusters_list, sil_coefs_tsne)
```

```python
plot_sse(n_clusters_list,inertias_tsne)
```

Treba uzeti u obzir da je prilikom rada sa t-SNE dimenzija za jedan manja u odnosu na PCA.
